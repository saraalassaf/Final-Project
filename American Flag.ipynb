```python
import turtle
import time

screen = turtle.getscreen()
screen.title("The American Flag")
screen.bgcolor("white")

turt = turtle.Turtle()
turt.speed(50)
turt.penup()

flag_height = 300
flag_width = 570

x1 = -300
y1 = 250

stripe_height = flag_height/13
stripe_width = flag_width
star_size = 10

def draw_rectangle(x, y, height, width, color):
    turt.goto(x,y)
    turt.pendown()
    turt.color(color)
    turt.begin_fill()
    turt.forward(width)
    turt.right(90)
    turt.forward(height)
    turt.right(90)
    turt.forward(width)
    turt.right(90)
    turt.forward(height)
    turt.right(90)
    turt.end_fill()
    turt.penup()
    
def shape_of_star(x, y, color,length):
    turt.goto(x,y)
    turt.setheading(0)
    turt.pendown()
    turt.begin_fill()
    turt.color(color)
    for i in range(0,5):
        turt.forward(length)
        turt.right(144)
        turt.forward(length)
        turt.right(144)
    turt.end_fill()
    turt.penup()
    
def draw_stripes():
    x = x1
    y = y1
    for i in range(6):
        for color in ["red", "white"]:
            draw_rectangle(x, y, stripe_height, stripe_width, color)
            y = y - stripe_height
            
    draw_rectangle(x, y, stripe_height, stripe_width, 'red')
    y = y - stripe_height

def draw_square():
    square_height = (7/13) * flag_height
    square_width = (0.4) * flag_width
    draw_rectangle(x1, y1, square_height, square_width, 'navy')
    
def stars1():
    dist_of_stars = 40
    dist_between_lines = stripe_height + 6
    y = 230
    for i in range(5):
        x = -290
        for j in range(6):
            shape_of_star(x, y, 'white', star_size)
            x = x + dist_of_stars
        y = y - dist_between_lines
        
def stars_five():
    dist_of_stars = 40
    dist_between_lines = stripe_height + 6
    y = 215
    for i in range(4):
        x = -270
        for j in range(5):
            shape_of_star(x, y, 'white', star_size)
            x = x + dist_of_stars
        y = y - dist_between_lines

draw_stripes()
draw_square()
stars1()
stars_five()
turt.hideturtle()
screen.mainloop()
```
